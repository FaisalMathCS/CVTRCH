## Tensor Operations

```python
x = torch.tensor(([1, 2, 3],
                 [4, 5, 6]))
```

```python
x * 10
```

```python
x + 10
```

```python
y = x.add(10)
```

```python
y
```

```python
y.shape
```

```python
y.view((3, 2))
```

```python
y = y.view((3, 2))
```

```python
x @ y
```

```python
y @ x
```

```python
x = torch.tensor([[2., -1.], [1., 1.]], requires_grad=True)
print(x)
```

```python
out = x.pow(2).sum()
```

```python
out.backward()
```

```python
x.grad
```
